# Outlier Detection and Removal

 Use visualization or logic to remove extreme values in Weekly_Sales or others. To clean the Weekly_Sales data, we use the Interquartile Range (IQR) method to remove outliers. Here's how it works:
Calculate Quartiles (Q1 and Q3):
Q1 is the 25th percentile, and Q3 is the 75th percentile of the data.
Compute IQR:
The IQR is the difference between Q3 and Q1:
IQR = Q3 - Q1.
Set Boundaries for Outliers:
Outliers are values outside the range:
Lower Bound: Q1 - 1.5 * IQR
Upper Bound: Q3 + 1.5 * IQR.
Remove Outliers:
Data points outside this range are removed:

```python
Q1 = df['Weekly_Sales'].quantile(0.25)
Q3 = df['Weekly_Sales'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers
df = df[(df['Weekly_Sales'] >= Q1 - 1.5*IQR) & (df['Weekly_Sales'] <= Q3 + 1.5*IQR)]
```

You can also remove outliers:



```python
q1 = df['Weekly_Sales'].quantile(0.25)
q3 = df['Weekly_Sales'].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df = df[(df['Weekly_Sales'] >= lower) & (df['Weekly_Sales'] <= upper)]
```

# Data Transformation (Normalize / Scale)

Normalization: Values are scaled to a fixed range, typically [0, 1].
Standardization: Values are rescaled to have a mean of 0 and a standard deviation of 1.
Effect on Distribution:

Normalization: Compresses or stretches the data to fit within the specified range, potentially altering the original distribution.
Standardization: Preserves the shape of the original distribution but changes the scale.
Use Cases:

Normalization: Suitable for distance-based algorithms, like k-nearest neighbors and neural networks. Scale numerical columns (like Temperature, CPI) for better model performance

# Exploration of this normalization section

Load the dataset: The dataset is loaded from the file.
Separate numeric and non-numeric columns: The numeric columns are separated for scaling, and the non-numeric columns are left unchanged.

# Standardization

Encodes categorical features:
Converts them into one-hot encoded dummy variables (so they can be used numerically).
Standardizes the numeric features:
Scales all numeric features to have zero mean and unit variance. This is crucial for PCA.
Applies PCA:
Reduces the high-dimensional dataset into fewer dimensions (up to 15 or less if fewer features exist).
PCA transforms the dataset into uncorrelated "principal components" that capture the most variance.
Prints PCA results:
Shows the shape of the PCA-transformed data and the first five rows of that transformed version.
Plots:
Left plot: Visualizes the original data using the first two numeric features.
Right plot: Visualizes the PCA-transformed data using its first two principal components.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Walmart dataset
data = pd.read_csv("C:\\Users\\National\\OneDrive\\Desktop\\Walmart_Sales.csv")

# Handle missing values (if any)
data.fillna(data.mean(numeric_only=True), inplace=True)

# Separate categorical and numeric columns
cat_features = [feature for feature in data.columns if data[feature].dtype == 'O']
numeric_features = [feature for feature in data.columns if data[feature].dtype != 'O']

# Convert categorical columns using pd.get_dummies
data = pd.get_dummies(data, columns=cat_features)

# Standardize the numeric columns
scaler = StandardScaler()
data[numeric_features] = scaler.fit_transform(data[numeric_features].values)

# Apply PCA
n_components = min(15, data.shape[1])  # Prevent requesting more components than available
pca = PCA(n_components=n_components)
data_pca = pca.fit_transform(data)

# Check the shape of the PCA data
print("PCA shape:", data_pca.shape)
print("First 5 PCA-transformed rows:\n", data_pca[:5])

# Plot the original data (first two numeric features)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(data[numeric_features[0]], data[numeric_features[1]], alpha=0.5)
plt.title('Original Data')
plt.xlabel(numeric_features[0])
plt.ylabel(numeric_features[1])

# Plot the PCA transformed data
plt.subplot(1, 2, 2)
plt.scatter(data_pca[:, 0], data_pca[:, 1], alpha=0.5)
plt.title('PCA Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```
